# strike rate

```python
# Group by batter
batsman_stats = ball.groupby('batter').agg({
    'batsman_runs': 'sum',
    'is_extra': lambda x: (x == 0).sum()
}).reset_index()

# Rename columns
batsman_stats.columns = ['batter', 'total_runs', 'total_balls']
print(batsman_stats)
```

```python
batsman_stats['strike_rate'] = (
    batsman_stats['total_runs'] / batsman_stats['total_balls']
) * 100
```

```python
filtered = batsman_stats[batsman_stats['total_balls'] >= 200]
```

```python
top_5_sr = filtered.sort_values(
    by='strike_rate',
    ascending=False
).head(10)

top_5_sr
```

# Boundary percentage 

```python
# total valid  balls
total_balls = ball[ball['is_extra'] == 0].groupby('batter').size().reset_index(name='total_balls')
print(total_balls)
```

```python
 # Boundary balls (4 or 6)
boundary = ball[
    (ball['batsman_runs'] == 4) | (ball['batsman_runs'] == 6)
]
boundary_count = boundary.groupby('batter').size().reset_index(name='boundary_balls')

# Merge
boundary_stats = total_balls.merge(boundary_count, on='batter', how='left')

# Fill NaN with 0
boundary_stats['boundary_balls'] = boundary_stats['boundary_balls'].fillna(0)

# Boundary percentage
boundary_stats['boundary_percentage'] = (
    boundary_stats['boundary_balls'] / boundary_stats['total_balls']
) * 100

# Apply 200 ball filter

filtered = boundary_stats[boundary_stats['total_balls'] >= 200]

# Top 5
filtered.sort_values(by='boundary_percentage', ascending=False).head(5)
```

# Average run per match

```python
run_per_match=ball.groupby(['batter','match_no'])['batsman_runs'].sum().reset_index()
print(run_per_match)
```

```python
batsman_avg=run_per_match.groupby('batter').agg({
    'batsman_runs':'sum',
    'match_no':'nunique'}).reset_index()
batsman_avg.columns = ['batter', 'total_runs', 'matches_played']
print(batsman_avg)
```

```python
batsman_avg['avg_runs_per_match'] = (
    batsman_avg['total_runs'] / batsman_avg['matches_played']
)
filtered = batsman_avg[batsman_avg['matches_played'] >= 5]
filtered.sort_values(
    by='avg_runs_per_match',
    ascending=False
).head(5)
```

# Economy Rate 
    (Condition: minimum 20 overs bowled)

```python
bowler_run=ball.groupby('bowler')['score'].size().reset_index(name='total_runs')
```

```python
bowler_ball=ball[ball['is_extra']==0].groupby('bowler').size().reset_index(name='total_balls')
```

```python
bowler_stat=bowler_run.merge(bowler_ball, on ='bowler')
```

```python
bowler_stat['economy'] = (
    bowler_stat['total_runs'] / bowler_stat['total_balls']
) * 6
```

```python
filtered = bowler_stat[bowler_stat['total_balls'] >= 120]
filtered.sort_values(
    by='economy',
    ascending=True
).head(10)
```

# Team Win Count & Win Percentage

```python
team1=match['team1']
team2=match['team2']
allteam=pd.concat([team1,team2])
matches_played = allteam.value_counts().reset_index()
matches_played.columns = ['team', 'matches_played']
```

```python
wins=match['winner'].value_counts().reset_index()
wins.columns = ['team', 'wins']
```

```python
team_stat=matches_played.merge(wins,on='team', how='left')
team_stat['wins'] = team_stat['wins'].fillna(0)
```

```python
team_stat['win_percentage'] = (
    team_stat['wins'] / team_stat['matches_played']
) * 100

team_stat.sort_values(
    by='win_percentage',
    ascending=False
)
```

# Venue-wise Average First Innings Score

```python
first_innings=ball[ball['inningno']==1]
```

```python
first_innings_score = first_innings.groupby('match_no')['score'].sum().reset_index()
first_innings_score.columns = ['match_no', 'first_innings_total']
```

```python
venue_data = first_innings_score.merge(
    match[['match_number', 'venue']],
    left_on='match_no',
    right_on='match_number'
)
```

```python
venue_avg = venue_data.groupby('venue')['first_innings_total'] \
    .mean() \
    .reset_index(name='avg_first_innings_score')

venue_avg.sort_values(
    by='avg_first_innings_score',
    ascending=False
)
```
